```python
import data_patterns
import pandas as pd
import numpy as np
filename = 'dic.txt'
output = 'dict.txt'
```

```python
df = pd.read_fwf(filename)
df = df.iloc[:,0].str.split('\t', 4, expand=True)
df = df.fillna(value=' ')
df
```

```python
with open(output, 'w') as f:
    f.writelines(
        '\t'.join(row) + '\n'
            for _, row in df.iterrows())
```
